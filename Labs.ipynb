# Student Performance Analysis

This notebook performs data quality assessment, preprocessing, exploratory data analysis (EDA), and builds a linear regression model to predict student performance index based on various factors.

## Import Libraries

Import necessary libraries for data manipulation, visualization, and machine learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Enable inline plotting for Jupyter
%matplotlib inline
```

## Load and Explore Dataset

Load the StudentPerformance.csv dataset and perform initial exploration including head, info, and descriptive statistics.

```python
df = pd.read_csv("StudentPerformance.csv")

print("First 5 rows of the dataset:")
print(df.head())
print("\nDataset information:")
print(df.info())
print("\nDescriptive statistics:")
print(df.describe())
```

## Data Preprocessing

Convert categorical 'Extracurricular Activities' column to numerical (Yes=1, No=0).

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes': 1, 'No': 0})
print("Preprocessing completed. Categorical column converted to numerical.")
```

## Exploratory Data Analysis (EDA)

Visualize the data using pairplot, histogram, and correlation heatmap.

```python
# Pairplot to visualize relationships between variables
sns.pairplot(df)
plt.title('Pairplot of All Variables')
plt.show()
```

```python
# Histogram of Performance Index
sns.histplot(df['Performance Index'])
plt.title('Distribution of Performance Index')
plt.xlabel('Performance Index')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Model Training

Split the data into training and testing sets, then train a linear regression model.

```python
# Prepare features and target
X = df.drop('Performance Index', axis=1)
y = df['Performance Index']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)

# Train model
lm = LinearRegression()
lm.fit(X_train, y_train)

print(f"Model intercept: {lm.intercept_}")
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
print("\nModel coefficients:")
print(coeff_df)
```

## Model Evaluation

Make predictions and evaluate the model using scatter plot, residuals histogram, and error metrics.

```python
# Make predictions
predictions = lm.predict(X_test)

# Scatter plot of actual vs predicted
plt.scatter(y_test, predictions)
plt.xlabel("Actual Performance Index")
plt.ylabel("Predicted Performance Index")
plt.title('Actual vs Predicted Performance Index')
plt.show()
```

```python
# Residuals histogram
sns.histplot((y_test - predictions), bins=50)
plt.title('Residuals Distribution')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
# Evaluation metrics
print('Model Evaluation Metrics:')
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
